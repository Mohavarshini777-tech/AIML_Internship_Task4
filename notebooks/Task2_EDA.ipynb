```python
# Step 1: Import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style="whitegrid")

print("Libraries imported successfully!")
```

```python
# Step 2: Load the Titanic dataset

file_path = "../data/titanic.csv"   # path from notebooks folder
df = pd.read_csv(file_path)

# Show first 5 rows
df.head()
```

```python
# Step 3: Summary Statistics

print("---- Numerical Summary ----")
display(df.describe().T)   # summary for numeric columns

print("\n---- Categorical Summary ----")
display(df.describe(include='object').T)   # summary for object/categorical columns
```

```python
# Step 4: Missing Values Check

print("Missing values in each column:")
df.isnull().sum()
```

```python
# Step 5: Histograms for numerical columns

df.hist(figsize=(12, 8), edgecolor='black')
plt.suptitle("Histograms of Numerical Features", fontsize=16)
plt.show()
```

```python
# Step 6: Boxplots for numerical columns

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(y=df['Age'])
plt.title("Boxplot: Age")

plt.subplot(1, 2, 2)
sns.boxplot(y=df['Fare'])
plt.title("Boxplot: Fare")

plt.tight_layout()
plt.show()
```

```python
# Step 7: Countplots for categorical features

plt.figure(figsize=(15, 5))

# Sex
plt.subplot(1, 3, 1)
sns.countplot(x=df['Sex'])
plt.title("Gender Distribution")

# Pclass
plt.subplot(1, 3, 2)
sns.countplot(x=df['Pclass'])
plt.title("Passenger Class Distribution")

# Survived
plt.subplot(1, 3, 3)
sns.countplot(x=df['Survived'])
plt.title("Survival Distribution")

plt.tight_layout()
plt.show()
```

```python
# Step 8: Correlation Heatmap

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Step 9: Pairplot for selected features

sns.pairplot(
    df[['Age', 'Fare', 'Pclass', 'Survived']], 
    hue='Survived',
    diag_kind='kde'
)
plt.show()
```
